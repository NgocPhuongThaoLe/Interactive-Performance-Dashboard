**Exercise 1:** Write a function count_vowels(s) that counts the number of vowels (a, e, i, o, u) in a given string 'This Is Just A String That We Use To Practice'

**Objetives:** Practice Problem Solving, String Manipulating, Loop, Define Function

**Input**: 'This Is Just A String That We Use To Practice'

**Output**: 13

```python
#input:string
#output: counts the number of vowels
#dÃ¹ng vÃ²ng láº·p kÃ¨m Ä‘iá»u kiá»‡n duyá»‡t tá»«ng chá»¯ cÃ¡i náº¿u in aeiou thÃ¬ cáº­p nháº­t count
#viáº¿t khung hÃ m :
def count_vowels(s):
  vowels = "aeiou"
  count = 0
  for char in s.lower():
    if char in vowels:
      count += 1
  return count
count_vowels('This Is Just A String That We Use To Practice')
```
